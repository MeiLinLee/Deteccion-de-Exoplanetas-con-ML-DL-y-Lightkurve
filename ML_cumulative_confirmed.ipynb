### Important Functions

```python
# Metricas de Machine Learning

def evaluation(y_true, y_pred):
    
# Print exactitud, exhaustividad, F1, precisión
    print('Evaluation Metrics:')
    print(f"Accuracy: {metrics.accuracy_score(y_test, y_pred)}")
    print(f"Recall: {metrics.recall_score(y_test, y_pred)}")
    print(f"F1 Score: {metrics.f1_score(y_test, y_pred)}")
    print(f"Precision: {metrics.precision_score(y_test, y_pred)}")
    
# Print Confusion Matrix
    print('\nConfusion Matrix:')
    print(' TN,  FP, FN, TP')
    print(confusion_matrix(y_true, y_pred).ravel())
    
# Mejores parametros para gridsearch
def print_results(results):
    print('Best Parameters: {}\n'.format(results.best_params_)) 
    
```

```python
features = df.drop(columns=['ExoplanetCandidate','ExoplanetConfirmed']) #elimina las columnas para el dataframe a utilizar en ML
target = df.ExoplanetConfirmed #determina la columna a buscar para el ML
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, random_state=2, test_size=.20) #divide el dataframe en datos de entrenamiento y de testeo 70% 30%
```

```python
# revisar los % de entrenamiento y testeo
for dataset in [y_train, y_test]:
    print(round(len(dataset)/len(target), 2))
```

```python
def fit_ev(modelo): 
    modelo.fit(X_train, y_train) # fitea el modelo
    y_pred = modelo.predict(X_test) # Predice en los datos de testeo
    evaluation(y_test, y_pred) # Evalua el modelo con las metricas de rendimiento
    return y_pred
```

```python
def model(classifier, n_neighbors, C): #
    modelo = None
    match classifier:
        #case "SVM": #se omite SVM porque es muy lento 
            #modelo = SVC(kernel = kernel)
        case "Naive Bayes":
            modelo = GaussianNB()
        case "KNN":
            modelo = KNeighborsClassifier(n_neighbors = n_neighbors, leaf_size=30, metric='canberra',weights='uniform')
        case "Tree":
            modelo = DecisionTreeClassifier()
        case "RandomForest":
            modelo = RandomForestClassifier(n_estimators=100, criterion='gini')
        case "Logistic":
            modelo = LogisticRegression(C = C)
    fit_ev(modelo)
```

```python
params = {"classifier" : widgets.Dropdown(options = ["Naive Bayes", "KNN", "Tree", "RandomForest", "Logistic"]),
#"kernel" : widgets.Dropdown(options= ["linear", "poly", "rbf", "sigmoid"]),
"n_neighbors" :widgets.IntSlider(min = 1, max = 20, value = 10),
"C" :  widgets.FloatSlider(min = 0, max = 10, value = 1, step = 0.1)}
```

```python
models_widgets = widgets.interact(model, **params) #permite interactuar con ML con un widget slider, Treem cambia de resultado aun cambiando variables que no deberian afectarle (bug?)
```

## Models

```python
rendimiento_modelos = pd.DataFrame()

def agregar_resultados(df, model_name, y_pred):
    return df.append({"Model": model_name, "Accuracy": metrics.accuracy_score(y_test, y_pred), "Recall": metrics.recall_score(y_test, y_pred), "F1 Score": metrics.f1_score(y_test, y_pred), "Precision": metrics.precision_score(y_test, y_pred)}, ignore_index=True)
```

```python
# Logistic Regression Model
lr = LogisticRegression(C=100, max_iter=2000, class_weight='balanced')

y_pred = fit_ev(lr)
rendimiento_modelos = agregar_resultados(rendimiento_modelos, "Regresion Logistica", y_pred)
```

```python
naive_bayes = GaussianNB()
y_pred = fit_ev(naive_bayes)
rendimiento_modelos = agregar_resultados(rendimiento_modelos, "Naive Bayes", y_pred)
```

```python
knn = KNeighborsClassifier(n_neighbors = 10, leaf_size=30, metric='canberra',weights='uniform')

y_pred = fit_ev(knn) 
rendimiento_modelos = agregar_resultados(rendimiento_modelos, "KNN", y_pred)
```

```python
tree = DecisionTreeClassifier()

y_pred = fit_ev(tree)
rendimiento_modelos = agregar_resultados(rendimiento_modelos, "Arbol de decision", y_pred)
```

```python
forest = RandomForestClassifier(n_estimators=100, criterion='gini')

fit_ev(forest)
rendimiento_modelos = agregar_resultados(rendimiento_modelos, "Bosque Aleatorio", y_pred)
```

```python
forest = RandomForestClassifier(n_estimators=10000, criterion='entropy') #puede ser overfitted numero estimadores muy grande vs numero datos

y_pred = fit_ev(forest)
```

```python
parameters = {
    'n_estimators': [5, 50, 100, 150, 200],
    'max_depth': list(range(1, 11)),
    'criterion':['gini','entropy'],
    'max_features': list(range(1,20)),
    'oob_score':[False,True],
} #parametros para gridsearch
```

```python
grid = GridSearchCV(forest, parameters, cv=5, verbose=1, n_jobs=-1) #mejora el modelo de forest

y_pred = fit_ev(grid)
print_results(grid)
rendimiento_modelos = agregar_resultados(rendimiento_modelos, "grid", y_pred)
```
